## 2. Define en python vectores y funciones
<p align="justify">
Debes generar vectores que representan intervalos de tiempo o de una variable.
Debes ingresar las funciones matemáticas y evaluarlas en los vectores generados.

$$
f(x) =  e^{-x^2}
$$

$$
f(z) =  ysin(x)
$$

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
valores_x = np.linspace(-5, 5, 100)  # Rango de -5 a 5 con 100 puntos

# Definir las funciones matemáticas
def f1(x):
    return np.exp(-x**2)

def f2(x, y):
    return y * np.sin(x)

# Evaluar las funciones en los vectores generados
valores_y1 = f1(valores_x)
valores_y2 = f2(valores_x, 2)  # y=2 para la segunda función, puedes ajustar este valor, sera la amplitud de la funcion seno

# Graficar las funciones
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(valores_x, valores_y1, label='$e^{-x^2}$', color='blue')
plt.title('Función $e^{-x^2}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(valores_x, valores_y2, label='$y \cdot sin(x)$', color='red')
plt.title('Función $y \cdot sin(x)$ con y=2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
valores_x = np.linspace(-5, 5, 100)  # Rango de -5 a 5 con 100 puntos

# Definir las funciones matemáticas
def f1(x):
    return np.exp(-x**2)

def f2(x, y):
    return y * np.sin(x)

# Evaluar las funciones en los vectores generados
valores_y1 = f1(valores_x)
valores_y2 = f2(valores_x, 2)  # y=2 para la segunda función, puedes ajustar este valor

# Generar ruido gaussiano

ruido_gau = np.random.normal(0, 0.1, len(valores_x))  # Media 0, Desviación estándar 0.1, se pueden editar

# Agregar ruido a las señales
ruido_y1 = valores_y1 + ruido_gau
ruido_y2 = valores_y2 + ruido_gau

# Graficar las señales originales y las señales con ruido
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(valores_x, valores_y1, label='Función original', color='blue')
plt.plot(valores_x, ruido_y1, label='Función con ruido', linestyle='--', color='orange')
plt.title('Función $e^{-x^2}$ con ruido gaussiano')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(valores_x, valores_y2, label='Función original', color='red')
plt.plot(valores_x, ruido_y2, label='Función con ruido', linestyle='--', color='green')
plt.title('Función $y \cdot sin(x)$ con ruido gaussiano')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
